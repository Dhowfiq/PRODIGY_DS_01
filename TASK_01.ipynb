**IMPORTING LIBRARIES:**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**LOADING THE DATA:**

```python
df = pd.read_csv('/content/ds_salaries.csv')

# Display the first few rows to understand the structure of the data
df.head()
print(df.head())
```

**VISUALISING THE DISTRIBUTION FOR CATEGORICAL VARIABLE**:

```python
import matplotlib.pyplot as plt


# Plotting the histogram for salary distribution
plt.figure(figsize=(10, 6))
# Replace 'salary_in_usd' with the actual name of your salary column if it's different
plt.hist(df['salary_in_usd'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()

```
